```python
vid = cv2.VideoCapture("./inputdata/sample.mp4")
while vid.isOpened():
    ret,frame = vid.read()
    if ret:
        cv2.imshow('video',frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
vid.release()
cv2.destroyAllWindows()
```
